⭐ Exercise 1 — Create Your First Conv Layer


```python
import torch
import torch.nn as nn

conv_layer = nn.Conv2d(
    in_channels=1,  # Input channels (1 for grayscale, 3 for RGB)
    out_channels=4,  # Number of filters in this layer
    kernel_size=3 # Size of each filter (3x3)
)
print(conv_layer)
```

⭐ Exercise 2 — Pass a Fake Image Through the Layer


```python
image = torch.rand(1, 1, 28, 28)  # Batch=1, Channels=1, H=28, W=28
output = conv_layer(image)
output.shape
# note how after the conv layer, the size of the image changes. H=26, W=26
```

⭐ Exercise 3 — Try Different Padding


```python
conv_pad = nn.Conv2d(
    in_channels=1,  # Input channels (1 for grayscale, 3 for RGB)
    out_channels=4,  # Number of filters in this layer
    kernel_size=3, # Size of each filter (3x3)
    padding=1
)
output_pad = conv_pad(image)
output_pad.shape
# note: padding keeps the height/width the same.
```

⭐ Exercise 4 — Try Changing the Stride


```python
conv_stride = nn.Conv2d(1, 4, 3, stride=1)
output_stride = conv_stride(image)
output_stride.shape
# note: the output of the image size
```
